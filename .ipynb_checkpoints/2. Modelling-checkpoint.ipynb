##  Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score, mean_absolute_error
import math
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.pipeline import Pipeline
```

##  Load data

```python
df_train = pd.read_csv("diamons_train.csv", index_col = 0)
df_test = pd.read_csv("diamons_test.csv", index_col = 0)
```

```python
df_test.head()
```

```python
y = df_train["price"]
X = df_train[['carat', "cut", "depth"]]
```

```python
X_test = df_test[['carat', "cut", "depth"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Many estimators

```python
pipe = Pipeline(steps=[('estimator', LinearRegression())])
```

```python
# Add a dict of estimator and estimator related parameters in this list
params_grid = [{
                'estimator':[LinearRegression()],
                },
                {
                'estimator': [DecisionTreeRegressor()],
                'estimator__max_depth': [1,2,3,4,5],
                'estimator__max_features': [None, "auto", "sqrt", "log2"],
                },
                {'estimator':[RandomForestRegressor()],
                 'estimator__n_estimators':[50,100,200,300],
                 'estimator__max_depth':[4, 5, 6, None], 
                 'estimator__min_samples_split':[2,5,8], 
                 'estimator__min_samples_leaf':[1,3,5,8],
                },
                {'estimator':[KNeighborsRegressor()],
                 'estimator__n_neighbors':[3],
                }

              ]
```

```python
grid = GridSearchCV(pipe, params_grid, n_jobs=1, verbose = 2)
```

```python
%%time
res = grid.fit(X, y)
```

```python
grid.best_params_
```

```python
pd.DataFrame(grid.cv_results_)
```

```python
mse_scorer = make_scorer(mean_squared_error)
cross_validate(estimator=RandomForestRegressor(max_depth=None,
                        min_samples_split=5,min_samples_leaf = 1, n_estimators=300), 
               X=X, y=y, scoring=mse_scorer, cv=10, verbose = 2)
```

```python
best_final = res.best_estimator_
```

```python
predictedY_test2 =  best_final.predict(X_test)
```

```python
mse_scorer = make_scorer(mean_squared_error)
cross_validate(estimator=RandomForestRegressor(max_depth=None,
                        min_samples_split=5,min_samples_leaf = 1, n_estimators=300), 
               X=X_test, y=predictedY_test2, scoring=mse_scorer, cv=10, verbose = 2)
```

```python
comparing_test2 = pd.DataFrame(predictedY_test2, columns = ["price"])
comparing_test2.index.name = 'id'
```

```python
comparing_test2.to_csv("Submission_3variables.csv", header = True)
```
